```
t1 = "Ethics are built right into the ideals and objectives of the United Nations"

#no. of chars
print(len(t1))

#no. of words
t2 = t1.split(' ')
print(len(t2))

#long words: words that are more than 3 letters long
print([w for w in t2 if len(w)>3])

#capitalize words
print([w for w in t2 if w.istitle()])

#words that end with s
print([w for w in t2 if w.endswith('s')])
```
